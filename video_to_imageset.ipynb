```python
import pandas as pd
import cv2
import os

def create_image_set(video_path, tag, index):
    vidcap = cv2.VideoCapture(f"train_video/{video_path}")
    length = int(vidcap.get(cv2.CAP_PROP_FRAME_COUNT))
    delta = length // 10
    current_frame, count = 1, 1
    print(f"Creating image set for {video_path}")
    for i in range(length):
        success, image = vidcap.read(current_frame)
        if success and i == current_frame:
            image = cv2.resize(image, (224, 224), interpolation = cv2.INTER_AREA)
            cv2.imwrite(f"dataset/{tag}/image_{index}_{count}.jpg", image)
            current_frame += delta
            count += 1

    vidcap.release()


data = pd.read_csv('train.csv')
paths = set(data.labels) - {'cloud, sunset', 'cloud, water'}

os.mkdir('dataset')
for path in paths:
    os.mkdir(f"dataset/{path}")

for index, row in data.iterrows():
    if row.labels in paths:
        create_image_set(row.path, row.labels, index)
```
